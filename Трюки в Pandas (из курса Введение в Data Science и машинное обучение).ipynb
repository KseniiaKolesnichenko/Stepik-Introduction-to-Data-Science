```python
import pandas as pd
import numpy as np 
```

```python
movie=pd.read_csv('Data\movie_metadata.csv')
movie.head()
```

```python
genres=movie[['movie_title', 'genres']]
genres.head()
```

##### Итерация по столбцам 

```python
for i in genres:
    print (i)
```

```python
for i in genres.columns:
    print (i)
```

##### Итерация по значениям

```python
for i in genres.values:
    print (i)
```

```python
# Итерация по названиям фильмов
for i in genres.values:
    print (i[0])
```

```python
# Итерация по жанрам фильмов
for i in genres.values:
    print (i[1])
```

```python
for row in genres.values:
    for value in row:
        print(value)
```

```python
def reversator(value):
    return value[::-1]
```

```python
for row in genres.values:
    for value in row:
        print(reversator(value))
```

###### Итерация по строкам iterrows

```python
#Вывод индекса строки и самой строки
for row in genres.iterrows():
    print(row)    
```

```python
# Вывод названия фамльма и его жанра (series)
for i, row in genres.iterrows():
    print (row)
#    print(row.map(reversator))
```

```python
# Вывод названия фамльма и его жанра (series)
# map применяет функцию, которая в него передана, каждому элементу серии
for i, row in genres.iterrows():
    print(row.map(reversator))
```

##### Итерация по колонкам

```python
for col in genres.iteritems():
    print(col)
```

```python
#Убираем заголовок
for i, col in genres.iteritems():
    print(col)
```

```python
budget=movie[['budget', 'duration']]
budget.head()
```

метод applymap применяет функцию, которую ему передали к каждой ячейке датафрейма

```python
budget.applymap(lambda x: x + 1)
```

apply можно применять и к строкам, и к столбцам.

```python
budget.apply(np.mean) # то же, что budget.apply(np.mean, axis=0) выводит среднее по столбцам
```

```python
budget.apply(np.mean, axis=1) #выводит среднее по строке
```

Применение функции ко всему датафрейму наиболее оптимально и по записи и по производительности. 
Обычно нампаевские эрреи быстрее, чем пандосовские (пандас хранит в себе больше структур, у него больше оверхед). Но! 
производительность методов зависит также от объёма данных, при большом числе наблюдений использовать пандасовские методы выгоднее чем извлекать эррей и работать с ним

```python
budget.mean()+1
```
